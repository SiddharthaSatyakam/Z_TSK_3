```python
X=ndf['Price']
```

```python
Y=ndf.drop(['Price'],axis=1)
Y['Rooms']=df['Avg. Area Number of Rooms']
Y['Bedrooms']=df['Avg. Area Number of Bedrooms']
Y=Y.drop(['b_ratio'],axis=1)
Y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
xgb=XGBRegressor()
lr=LinearRegression()
tts=train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(Y,X, test_size=0.2, random_state=1)
```

```python
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
from sklearn import metrics
```

```python
r2score=metrics.r2_score(y_test,pred)
r2score
```

```python
mae=metrics.mean_absolute_error(y_test,pred)
print(mae)
```
